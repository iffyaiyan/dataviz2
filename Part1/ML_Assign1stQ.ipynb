```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split
from torchvision import datasets, transforms
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import seaborn as sns
from tqdm import tqdm
import random

# Set seeds for reproducibility
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed()

# Configuration
class Config:
    batch_size = 64
    epochs = 20
    learning_rate = 0.001
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    val_size = 0.1  # 10% of training data for validation

# Data loading and preprocessing
class MNISTDataLoader:
    def __init__(self, batch_size):
        self.batch_size = batch_size
        self.transform = transforms.Compose([
            transforms.ToTensor(),
            transforms.Normalize((0.1307,), (0.3081,))  # Mean and std of MNIST
        ])

    def load_data(self):
        # Load MNIST dataset
        train_dataset = datasets.MNIST(
            root='./data',
            train=True,
            download=True,
            transform=self.transform
        )

        # Split training data into train and validation sets
        train_size = int((1 - Config.val_size) * len(train_dataset))
        val_size = len(train_dataset) - train_size
        train_dataset, val_dataset = random_split(train_dataset, [train_size, val_size])

        test_dataset = datasets.MNIST(
            root='./data',
            train=False,
            download=True,
            transform=self.transform
        )

        # Create data loaders
        train_loader = DataLoader(
            train_dataset,
            batch_size=self.batch_size,
            shuffle=True
        )

        val_loader = DataLoader(
            val_dataset,
            batch_size=self.batch_size,
            shuffle=False
        )

        test_loader = DataLoader(
            test_dataset,
            batch_size=self.batch_size,
            shuffle=False
        )

        return train_loader, val_loader, test_loader

# Model definition - MLP with configurable architecture
class MLP(nn.Module):
    def __init__(self, layer_sizes, dropout_rate=0.2):
        super(MLP, self).__init__()

        layers = []
        for i in range(len(layer_sizes) - 1):
            layers.append(nn.Linear(layer_sizes[i], layer_sizes[i+1]))

            # Add ReLU and Dropout for all but the last layer
            if i < len(layer_sizes) - 2:
                layers.append(nn.ReLU())
                layers.append(nn.Dropout(dropout_rate))

        self.model = nn.Sequential(*layers)

    def forward(self, x):
        # Flatten the input
        x = x.view(x.size(0), -1)
        return self.model(x)

    def get_activation_maps(self, x, layer_index):
        """Get activation maps for a specific layer"""
        x = x.view(x.size(0), -1)

        # Go through layers up to the specified layer
        for i, layer in enumerate(self.model):
            x = layer(x)
            if i == layer_index:
                return x

        return x

# Trainer class
class ModelTrainer:
    def __init__(self, model, optimizer, criterion, device):
        self.model = model
        self.optimizer = optimizer
        self.criterion = criterion
        self.device = device
        self.train_losses = []
        self.val_losses = []
        self.train_accuracies = []
        self.val_accuracies = []

    def train_epoch(self, train_loader):
        self.model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        for inputs, targets in tqdm(train_loader, desc="Training"):
            inputs, targets = inputs.to(self.device), targets.to(self.device)

            # Zero the gradients
            self.optimizer.zero_grad()

            # Forward pass
            outputs = self.model(inputs)
            loss = self.criterion(outputs, targets)

            # Backward pass and optimize
            loss.backward()
            self.optimizer.step()

            # Statistics
            running_loss += loss.item() * inputs.size(0)
            _, predicted = torch.max(outputs, 1)
            total += targets.size(0)
            correct += (predicted == targets).sum().item()

        epoch_loss = running_loss / total
        epoch_acc = correct / total
        self.train_losses.append(epoch_loss)
        self.train_accuracies.append(epoch_acc)

        return epoch_loss, epoch_acc

    def validate(self, val_loader):
        self.model.eval()
        running_loss = 0.0
        correct = 0
        total = 0

        with torch.no_grad():
            for inputs, targets in tqdm(val_loader, desc="Validation"):
                inputs, targets = inputs.to(self.device), targets.to(self.device)

                # Forward pass
                outputs = self.model(inputs)
                loss = self.criterion(outputs, targets)

                # Statistics
                running_loss += loss.item() * inputs.size(0)
                _, predicted = torch.max(outputs, 1)
                total += targets.size(0)
                correct += (predicted == targets).sum().item()

        epoch_loss = running_loss / total
        epoch_acc = correct / total
        self.val_losses.append(epoch_loss)
        self.val_accuracies.append(epoch_acc)

        return epoch_loss, epoch_acc

    def train(self, train_loader, val_loader, epochs):
        best_val_loss = float('inf')
        best_model_state = None

        for epoch in range(epochs):
            train_loss, train_acc = self.train_epoch(train_loader)
            val_loss, val_acc = self.validate(val_loader)

            print(f"Epoch {epoch+1}/{epochs} - "
                  f"Train Loss: {train_loss:.4f} - Train Acc: {train_acc:.4f} - "
                  f"Val Loss: {val_loss:.4f} - Val Acc: {val_acc:.4f}")

            # Save best model
            if val_loss < best_val_loss:
                best_val_loss = val_loss
                best_model_state = self.model.state_dict().copy()

        # Load best model
        if best_model_state:
            self.model.load_state_dict(best_model_state)

        return self.train_losses, self.val_losses, self.train_accuracies, self.val_accuracies

    def evaluate(self, test_loader):
        self.model.eval()
        all_preds = []
        all_targets = []

        with torch.no_grad():
            for inputs, targets in tqdm(test_loader, desc="Testing"):
                inputs, targets = inputs.to(self.device), targets.to(self.device)

                # Forward pass
                outputs = self.model(inputs)
                _, predicted = torch.max(outputs, 1)

                all_preds.extend(predicted.cpu().numpy())
                all_targets.extend(targets.cpu().numpy())

        # Calculate metrics
        accuracy = accuracy_score(all_targets, all_preds)
        precision = precision_score(all_targets, all_preds, average='weighted')
        recall = recall_score(all_targets, all_preds, average='weighted')
        f1 = f1_score(all_targets, all_preds, average='weighted')

        # Confusion matrix
        cm = confusion_matrix(all_targets, all_preds)

        return {
            'accuracy': accuracy,
            'precision': precision,
            'recall': recall,
            'f1': f1,
            'confusion_matrix': cm,
            'predictions': all_preds,
            'targets': all_targets
        }

# Analyzer class for model interpretation
class ModelAnalyzer:
    def __init__(self, model, data_loader, device):
        self.model = model
        self.data_loader = data_loader
        self.device = device

    def get_misclassified_samples(self, limit=10):
        """Get misclassified examples"""
        self.model.eval()
        misclassified_inputs = []
        misclassified_targets = []
        misclassified_preds = []

        with torch.no_grad():
            for inputs, targets in self.data_loader:
                inputs, targets = inputs.to(self.device), targets.to(self.device)
                outputs = self.model(inputs)
                _, preds = torch.max(outputs, 1)

                # Find misclassified examples
                misclassified_idx = (preds != targets).nonzero(as_tuple=True)[0]

                for idx in misclassified_idx:
                    misclassified_inputs.append(inputs[idx].cpu())
                    misclassified_targets.append(targets[idx].item())
                    misclassified_preds.append(preds[idx].item())

                    if len(misclassified_inputs) >= limit:
                        break

                if len(misclassified_inputs) >= limit:
                    break

        return misclassified_inputs, misclassified_targets, misclassified_preds

    def get_correctly_classified_samples(self, limit=10):
        """Get correctly classified examples"""
        self.model.eval()
        correct_inputs = []
        correct_targets = []

        with torch.no_grad():
            for inputs, targets in self.data_loader:
                inputs, targets = inputs.to(self.device), targets.to(self.device)
                outputs = self.model(inputs)
                _, preds = torch.max(outputs, 1)

                # Find correctly classified examples
                correct_idx = (preds == targets).nonzero(as_tuple=True)[0]

                for idx in correct_idx:
                    correct_inputs.append(inputs[idx].cpu())
                    correct_targets.append(targets[idx].item())

                    if len(correct_inputs) >= limit:
                        break

                if len(correct_inputs) >= limit:
                    break

        return correct_inputs, correct_targets

    def visualize_layer_activations(self, input_image, layer_indices):
        """Visualize activations for specific layers"""
        self.model.eval()
        input_image = input_image.unsqueeze(0).to(self.device)

        activations = []
        for layer_idx in layer_indices:
            act = self.model.get_activation_maps(input_image, layer_idx)
            activations.append(act.cpu().detach().numpy()[0])

        return activations

    def plot_weight_distributions(self):
        """Plot distribution of weights in each layer"""
        fig, axes = plt.subplots(len(self.model.model) // 3 + 1, 1, figsize=(10, 12))
        layer_idx = 0

        for i, layer in enumerate(self.model.model):
            if isinstance(layer, nn.Linear):
                weights = layer.weight.data.cpu().numpy().flatten()
                axes[layer_idx].hist(weights, bins=50)
                axes[layer_idx].set_title(f'Layer {layer_idx+1} Weight Distribution')
                axes[layer_idx].set_xlabel('Weight Value')
                axes[layer_idx].set_ylabel('Frequency')
                layer_idx += 1

        plt.tight_layout()
        return fig

    def visualize_feature_importance(self, test_loader, num_samples=100):
        """Compute input feature importance by occlusion"""
        self.model.eval()

        # Get a batch of data
        images, labels = next(iter(test_loader))
        images = images[:num_samples].to(self.device)
        labels = labels[:num_samples].to(self.device)

        # Get baseline predictions
        with torch.no_grad():
            baseline_output = self.model(images)
            baseline_pred = torch.softmax(baseline_output, dim=1)

        # Create importance map
        importance_map = torch.zeros_like(images[0, 0]).to(self.device)

        # Define patch size
        patch_size = 4

        # Iterate over patches
        for i in range(0, 28, patch_size):
            for j in range(0, 28, patch_size):
                # Create masked images
                masked_images = images.clone()
                masked_images[:, :, i:i+patch_size, j:j+patch_size] = 0

                with torch.no_grad():
                    masked_output = self.model(masked_images)
                    masked_pred = torch.softmax(masked_output, dim=1)

                # Compute importance as change in confidence
                importance = torch.abs(baseline_pred - masked_pred).mean(dim=(0, 1))
                importance_map[i:i+patch_size, j:j+patch_size] = importance.cpu()

        return importance_map.cpu().numpy()



  # Visualization utilities (continued)
class Visualizer:
    @staticmethod
    def plot_training_curves(train_losses, val_losses, train_accs, val_accs):
        fig, axs = plt.subplots(1, 2, figsize=(15, 5))

        # Plot loss curves
        axs[0].plot(train_losses, label='Train Loss')
        axs[0].plot(val_losses, label='Validation Loss')
        axs[0].set_title('Loss Curves')
        axs[0].set_xlabel('Epoch')
        axs[0].set_ylabel('Loss')
        axs[0].legend()

        # Plot accuracy curves
        axs[1].plot(train_accs, label='Train Accuracy')
        axs[1].plot(val_accs, label='Validation Accuracy')
        axs[1].set_title('Accuracy Curves')
        axs[1].set_xlabel('Epoch')
        axs[1].set_ylabel('Accuracy')
        axs[1].legend()

        plt.tight_layout()
        return fig

    @staticmethod
    def plot_confusion_matrix(cm):
        plt.figure(figsize=(10, 8))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
        plt.xlabel('Predicted Label')
        plt.ylabel('True Label')
        plt.title('Confusion Matrix')
        return plt.gcf()

    @staticmethod
    def plot_misclassified_examples(inputs, targets, preds, num_examples=10):
        num_examples = min(num_examples, len(inputs))
        fig, axes = plt.subplots(2, 5, figsize=(15, 6))
        axes = axes.flatten()

        for i in range(num_examples):
            img = inputs[i].squeeze().numpy()
            ax = axes[i]
            ax.imshow(img, cmap='gray')
            ax.set_title(f'True: {targets[i]}, Pred: {preds[i]}')
            ax.axis('off')

        plt.tight_layout()
        return fig

    @staticmethod
    def plot_layer_activations(activations, layer_names, num_units_per_layer=10):
        num_layers = len(activations)
        fig, axes = plt.subplots(num_layers, 1, figsize=(15, 3*num_layers))

        if num_layers == 1:
            axes = [axes]

        for i, (act, name) in enumerate(zip(activations, layer_names)):
            # Select a subset of neurons to display
            if len(act) > num_units_per_layer:
                indices = np.linspace(0, len(act)-1, num_units_per_layer, dtype=int)
                act_subset = act[indices]
            else:
                act_subset = act

            # Plot activations as bar chart
            axes[i].bar(range(len(act_subset)), act_subset)
            axes[i].set_title(f'{name} Activations')
            axes[i].set_xlabel('Neuron')
            axes[i].set_ylabel('Activation')

        plt.tight_layout()
        return fig

    @staticmethod
    def plot_feature_importance(importance_map):
        plt.figure(figsize=(8, 8))
        plt.imshow(importance_map, cmap='hot')
        plt.colorbar(label='Importance')
        plt.title('Feature Importance Map')
        plt.tight_layout()
        return plt.gcf()

# Experiment manager class
class ExperimentManager:
    def __init__(self, config):
        self.config = config
        self.data_loader = MNISTDataLoader(config.batch_size)
        self.train_loader, self.val_loader, self.test_loader = self.data_loader.load_data()
        self.device = config.device
        self.results = {}

    def run_experiment(self, model_params, exp_name):
        print(f"\n--- Running experiment: {exp_name} ---")
        print(f"Parameters: {model_params}")

        # Create model based on parameters
        input_size = 28 * 28  # MNIST image size
        output_size = 10  # 10 digits

        # Build layer sizes list including input and output
        layer_sizes = [input_size] + model_params['hidden_layers'] + [output_size]

        # Create model
        model = MLP(layer_sizes, dropout_rate=model_params.get('dropout_rate', 0.2))
        model.to(self.device)

        # Create optimizer
        optimizer = optim.Adam(
            model.parameters(),
            lr=model_params.get('learning_rate', self.config.learning_rate),
            weight_decay=model_params.get('weight_decay', 0)
        )

        # Loss function
        criterion = nn.CrossEntropyLoss()

        # Create trainer
        trainer = ModelTrainer(model, optimizer, criterion, self.device)

        # Train model
        train_losses, val_losses, train_accs, val_accs = trainer.train(
            self.train_loader,
            self.val_loader,
            model_params.get('epochs', self.config.epochs)
        )

        # Evaluate on test set
        test_metrics = trainer.evaluate(self.test_loader)

        # Store results
        self.results[exp_name] = {
            'model': model,
            'model_params': model_params,
            'train_losses': train_losses,
            'val_losses': val_losses,
            'train_accs': train_accs,
            'val_accs': val_accs,
            'test_metrics': test_metrics
        }

        print(f"\n--- Results for {exp_name} ---")
        print(f"Test Accuracy: {test_metrics['accuracy']:.4f}")
        print(f"Test Precision: {test_metrics['precision']:.4f}")
        print(f"Test Recall: {test_metrics['recall']:.4f}")
        print(f"Test F1 Score: {test_metrics['f1']:.4f}")

        return self.results[exp_name]

    def analyze_best_model(self):
        # Find best model based on test accuracy
        best_exp = max(self.results.items(), key=lambda x: x[1]['test_metrics']['accuracy'])
        best_exp_name, best_exp_results = best_exp

        print(f"\n=== Analyzing Best Model: {best_exp_name} ===")
        print(f"Test Accuracy: {best_exp_results['test_metrics']['accuracy']:.4f}")

        # Get model
        model = best_exp_results['model']

        # Create analyzer
        analyzer = ModelAnalyzer(model, self.test_loader, self.device)

        # Get misclassified examples
        misclass_inputs, misclass_targets, misclass_preds = analyzer.get_misclassified_samples(limit=10)

        # Get correctly classified examples
        correct_inputs, correct_targets = analyzer.get_correctly_classified_samples(limit=10)

        # Visualize confusion matrix
        cm_fig = Visualizer.plot_confusion_matrix(best_exp_results['test_metrics']['confusion_matrix'])
        cm_fig.savefig('confusion_matrix.png')

        # Visualize misclassified examples
        misclass_fig = Visualizer.plot_misclassified_examples(misclass_inputs, misclass_targets, misclass_preds)
        misclass_fig.savefig('misclassified_examples.png')

        # Compute and visualize feature importance
        importance_map = analyzer.visualize_feature_importance(self.test_loader)
        imp_fig = Visualizer.plot_feature_importance(importance_map)
        imp_fig.savefig('feature_importance.png')

        # Visualize layer activations for correct and misclassified examples
        if len(misclass_inputs) > 0 and len(correct_inputs) > 0:
            # Define which layers to visualize (layer indices for Linear layers)
            linear_indices = [i for i, layer in enumerate(model.model) if isinstance(layer, nn.Linear)]

            # Get layer names
            layer_names = [f'Layer {i//3 + 1}' for i in linear_indices]

            # Visualize activations for a misclassified example
            misclass_acts = analyzer.visualize_layer_activations(misclass_inputs[0].to(self.device), linear_indices)
            misclass_act_fig = Visualizer.plot_layer_activations(misclass_acts, layer_names)
            misclass_act_fig.savefig('misclassified_activations.png')

            # Visualize activations for a correctly classified example
            correct_acts = analyzer.visualize_layer_activations(correct_inputs[0].to(self.device), linear_indices)
            correct_act_fig = Visualizer.plot_layer_activations(correct_acts, layer_names)
            correct_act_fig.savefig('correct_activations.png')

        # Visualize weight distributions
        weight_dist_fig = analyzer.plot_weight_distributions()
        weight_dist_fig.savefig('weight_distributions.png')

        # Plot training curves
        curves_fig = Visualizer.plot_training_curves(
            best_exp_results['train_losses'],
            best_exp_results['val_losses'],
            best_exp_results['train_accs'],
            best_exp_results['val_accs']
        )
        curves_fig.savefig('training_curves.png')

        return {
            'best_exp_name': best_exp_name,
            'test_metrics': best_exp_results['test_metrics'],
            'model': model,
            'figures': {
                'confusion_matrix': cm_fig,
                'misclassified_examples': misclass_fig,
                'feature_importance': imp_fig,
                'training_curves': curves_fig,
                'weight_distributions': weight_dist_fig
            }
        }

# Main script to run experiments
def main():
    config = Config()

    # Define experiments with different architectures
    experiments = {
        'small_mlp': {
            'hidden_layers': [128, 64],
            'dropout_rate': 0.2,
            'learning_rate': 0.001,
            'weight_decay': 0.0001,
            'epochs': 15
        },
        'medium_mlp': {
            'hidden_layers': [256, 128, 64],
            'dropout_rate': 0.3,
            'learning_rate': 0.001,
            'weight_decay': 0.0001,
            'epochs': 15
        },
        'large_mlp': {
            'hidden_layers': [512, 256, 128, 64],
            'dropout_rate': 0.4,
            'learning_rate': 0.001,
            'weight_decay': 0.0001,
            'epochs': 15
        },
        'wide_mlp': {
            'hidden_layers': [1024, 512],
            'dropout_rate': 0.5,
            'learning_rate': 0.001,
            'weight_decay': 0.0001,
            'epochs': 15
        },
        'deep_mlp': {
            'hidden_layers': [256, 256, 128, 128, 64],
            'dropout_rate': 0.3,
            'learning_rate': 0.001,
            'weight_decay': 0.0001,
            'epochs': 15
        },
    }

    # Create experiment manager
    manager = ExperimentManager(config)

    # Run all experiments
    for exp_name, params in experiments.items():
        manager.run_experiment(params, exp_name)

    # Analyze best model
    analysis_results = manager.analyze_best_model()

    print("\n=== Experiment Complete ===")
    print(f"Best Model: {analysis_results['best_exp_name']}")
    print(f"Test Accuracy: {analysis_results['test_metrics']['accuracy']:.4f}")

    # Show plots if needed
    plt.show()

# Run main function
if __name__ == "__main__":
    main()
```

```python
# Enhanced model interpretability and analysis (continued)
def advanced_analysis():
    config = Config()

    # Load data
    data_loader = MNISTDataLoader(config.batch_size)
    train_loader, val_loader, test_loader = data_loader.load_data()

    # Load the best model (you should replace this with loading your actual best model)
    input_size = 28 * 28
    hidden_layers = [512, 256, 128, 64]  # Example architecture
    layer_sizes = [input_size] + hidden_layers + [10]

    model = MLP(layer_sizes)
    # model.load_state_dict(torch.load('best_model.pth'))  # Uncomment to load saved model
    model.to(config.device)

    # Analysis of misclassifications
    analyzer = ModelAnalyzer(model, test_loader, config.device)

    # Get more misclassified examples for pattern analysis
    misclass_inputs, misclass_targets, misclass_preds = analyzer.get_misclassified_samples(limit=100)

    # Analyze which digits are most commonly confused
    confusion_pairs = [(t, p) for t, p in zip(misclass_targets, misclass_preds)]
    confusion_counts = {}
    for pair in confusion_pairs:
        if pair in confusion_counts:
            confusion_counts[pair] += 1
        else:
            confusion_counts[pair] = 1

    # Find most common confusion pairs
    top_confusions = sorted(confusion_counts.items(), key=lambda x: x[1], reverse=True)[:5]
    print("\n=== Most Common Confusion Pairs ===")
    for (true, pred), count in top_confusions:
        print(f"True: {true}, Predicted: {pred} - Count: {count}")

    # Analyze activations for common misclassification cases
    if len(misclass_inputs) > 0:
        # Find indices of specific misclassification cases
        idx_8_3 = next((i for i, (t, p) in enumerate(zip(misclass_targets, misclass_preds))
                        if t == 8 and p == 3), None)

        if idx_8_3 is not None:
            print("\n=== Analyzing 8 classified as 3 ===")
            # Visualize this specific example
            fig, ax = plt.subplots(1, 1, figsize=(5, 5))
            ax.imshow(misclass_inputs[idx_8_3].squeeze().numpy(), cmap='gray')
            ax.set_title(f'True: 8, Predicted: 3')
            ax.axis('off')
            fig.savefig('8_as_3_example.png')

            # Get layer activations for this example
            linear_indices = [i for i, layer in enumerate(model.model) if isinstance(layer, nn.Linear)]
            layer_names = [f'Layer {i//3 + 1}' for i in linear_indices]

            misclass_acts = analyzer.visualize_layer_activations(misclass_inputs[idx_8_3].to(config.device), linear_indices)
            misclass_act_fig = Visualizer.plot_layer_activations(misclass_acts, layer_names)
            misclass_act_fig.savefig('8_as_3_activations.png')

    # Analyze examples with high confidence but wrong prediction
    # This helps understand where the model is most confidently wrong
    high_confidence_errors = []
    with torch.no_grad():
        model.eval()
        for inputs, targets in test_loader:
            inputs, targets = inputs.to(config.device), targets.to(config.device)
            outputs = model(inputs)
            probs = torch.nn.functional.softmax(outputs, dim=1)

            # Get predicted class and its probability
            max_probs, preds = torch.max(probs, dim=1)

            # Find where predictions are wrong
            errors = (preds != targets)

            # Get indices of high confidence errors
            for i in range(len(inputs)):
                if errors[i] and max_probs[i] > 0.9:  # High confidence threshold
                    high_confidence_errors.append({
                        'input': inputs[i].cpu(),
                        'true': targets[i].item(),
                        'pred': preds[i].item(),
                        'confidence': max_probs[i].item()
                    })

            if len(high_confidence_errors) >= 10:
                break

    if high_confidence_errors:
        print("\n=== High Confidence Errors ===")
        for i, err in enumerate(high_confidence_errors[:5]):
            print(f"Example {i+1}: True={err['true']}, Pred={err['pred']}, Confidence={err['confidence']:.4f}")

        # Visualize high confidence errors
        fig, axes = plt.subplots(1, min(5, len(high_confidence_errors)), figsize=(15, 3))
        if len(high_confidence_errors) == 1:
            axes = [axes]

        for i in range(min(5, len(high_confidence_errors))):
            axes[i].imshow(high_confidence_errors[i]['input'].squeeze().numpy(), cmap='gray')
            axes[i].set_title(f"T:{high_confidence_errors[i]['true']}, P:{high_confidence_errors[i]['pred']}\nConf:{high_confidence_errors[i]['confidence']:.2f}")
            axes[i].axis('off')

        plt.tight_layout()
        fig.savefig('high_confidence_errors.png')

    # Analyze decision boundaries by generating synthetic examples
    def generate_interpolated_images(img1, img2, steps=10):
        """Generate interpolated images between two examples"""
        interpolations = []
        for alpha in np.linspace(0, 1, steps):
            interp = img1 * (1 - alpha) + img2 * alpha
            interpolations.append(interp)
        return interpolations

    # Find examples of digits that are commonly confused (e.g., 4 and 9)
    digit_examples = {}
    for i in range(10):
        digit_examples[i] = None

    # Get one example of each digit
    for inputs, targets in test_loader:
        for i in range(len(inputs)):
            digit = targets[i].item()
            if digit_examples[digit] is None:
                digit_examples[digit] = inputs[i]

        if all(ex is not None for ex in digit_examples.values()):
            break

    # Create interpolations between commonly confused digits
    confusion_pairs_to_analyze = [(4, 9), (3, 8), (5, 8)]

    for d1, d2 in confusion_pairs_to_analyze:
        if digit_examples[d1] is not None and digit_examples[d2] is not None:
            print(f"\n=== Analyzing interpolation between {d1} and {d2} ===")

            # Generate interpolated images
            interpolations = generate_interpolated_images(
                digit_examples[d1].to(config.device),
                digit_examples[d2].to(config.device),
                steps=10
            )

            # Get predictions for interpolated images
            model.eval()
            with torch.no_grad():
                interp_preds = []
                interp_probs = []

                for img in interpolations:
                    output = model(img.unsqueeze(0))
                    prob = torch.nn.functional.softmax(output, dim=1)
                    _, pred = torch.max(prob, dim=1)
                    interp_preds.append(pred.item())
                    interp_probs.append(prob.squeeze().cpu().numpy())

            # Visualize interpolations and predictions
            fig, axes = plt.subplots(2, 5, figsize=(15, 6))
            axes = axes.flatten()

            for i in range(10):
                axes[i].imshow(interpolations[i].squeeze().cpu().numpy(), cmap='gray')
                axes[i].set_title(f"Pred: {interp_preds[i]}")
                axes[i].axis('off')

            plt.tight_layout()
            fig.savefig(f'interpolation_{d1}_{d2}.png')

            # Plot probability distributions for both classes during interpolation
            fig, ax = plt.subplots(figsize=(10, 6))

            interp_steps = np.arange(10)
            for digit in range(10):
                probs = [p[digit] for p in interp_probs]
                ax.plot(interp_steps, probs, 'o-', label=f'Digit {digit}')

            ax.set_xlabel('Interpolation Step')
            ax.set_ylabel('Probability')
            ax.set_title(f'Class Probabilities During Interpolation from {d1} to {d2}')
            ax.legend()

            plt.tight_layout()
            fig.savefig(f'interp_probs_{d1}_{d2}.png')

    # Class weight analysis
    print("\n=== Class Weight Analysis ===")

    # Get weights from the output layer
    output_layer = model.model[-1]  # Last layer
    output_weights = output_layer.weight.data.cpu().numpy()

    # For each digit, find which features are most important
    for digit in range(10):
        # Get weights for this digit
        digit_weights = output_weights[digit]

        # Find top features (highest absolute weights)
        top_feature_indices = np.argsort(np.abs(digit_weights))[-10:]

        # Create a feature importance map
        feature_map = np.zeros((28, 28))
        for idx in range(len(digit_weights)):
            i, j = idx // 28, idx % 28
            feature_map[i, j] = digit_weights[idx]

        # Plot feature importance map
        fig, ax = plt.subplots(figsize=(6, 6))
        im = ax.imshow(feature_map, cmap='coolwarm')
        ax.set_title(f'Feature Importance for Digit {digit}')
        plt.colorbar(im, ax=ax)
        fig.savefig(f'feature_map_digit_{digit}.png')

    # Analyze distribution of model confidence across all test data
    confidences = []
    correct_confidences = []
    incorrect_confidences = []

    model.eval()
    with torch.no_grad():
        for inputs, targets in test_loader:
            inputs, targets = inputs.to(config.device), targets.to(config.device)
            outputs = model(inputs)
            probs = torch.nn.functional.softmax(outputs, dim=1)

            # Get predicted class and its probability
            max_probs, preds = torch.max(probs, dim=1)

            # Store confidences
            for i in range(len(inputs)):
                confidences.append(max_probs[i].item())
                if preds[i] == targets[i]:
                    correct_confidences.append(max_probs[i].item())
                else:
                    incorrect_confidences.append(max_probs[i].item())

    # Plot confidence distributions
    fig, ax = plt.subplots(figsize=(10, 6))

    bins = np.linspace(0, 1, 20)
    ax.hist(correct_confidences, bins=bins, alpha=0.5, color='green', label='Correct')
    ax.hist(incorrect_confidences, bins=bins, alpha=0.5, color='red', label='Incorrect')

    ax.set_xlabel('Confidence')
    ax.set_ylabel('Count')
    ax.set_title('Distribution of Model Confidence')
    ax.legend()

    plt.tight_layout()
    fig.savefig('confidence_distribution.png')

    return {
        'top_confusions': top_confusions,
        'high_confidence_errors': high_confidence_errors
    }




# Training with sample weighting (continued)
def weighted_training_experiment():
    """
    Experiment with weighted training to give more importance to
    difficult examples or commonly confused classes
    """
    config = Config()

    # Load data
    data_loader = MNISTDataLoader(config.batch_size)
    train_loader, val_loader, test_loader = data_loader.load_data()

    # Define model architecture
    input_size = 28 * 28
    hidden_layers = [512, 256, 128, 64]
    layer_sizes = [input_size] + hidden_layers + [10]

    model = MLP(layer_sizes)
    model.to(config.device)

    # First, find commonly confused classes
    # Run a quick pre-training to identify difficult examples
    pre_model = MLP([input_size, 256, 128, 10])
    pre_model.to(config.device)
    optimizer = optim.Adam(pre_model.parameters(), lr=0.001)
    criterion = nn.CrossEntropyLoss()

    # Quick training
    pre_model.train()
    for epoch in range(3):
        for inputs, targets in train_loader:
            inputs, targets = inputs.to(config.device), targets.to(config.device)
            optimizer.zero_grad()
            outputs = pre_model(inputs)
            loss = criterion(outputs, targets)
            loss.backward()
            optimizer.step()

    # Evaluate and find difficult examples
    pre_model.eval()
    class_confusion = np.zeros((10, 10))
    with torch.no_grad():
        for inputs, targets in train_loader:
            inputs, targets = inputs.to(config.device), targets.to(config.device)
            outputs = pre_model(inputs)
            _, preds = torch.max(outputs, 1)

            # Update confusion matrix
            for t, p in zip(targets.cpu().numpy(), preds.cpu().numpy()):
                class_confusion[t, p] += 1

    # Normalize confusion matrix by row (true class)
    class_confusion_norm = class_confusion / (class_confusion.sum(axis=1, keepdims=True) + 1e-10)

    # Define class weights based on confusion matrix diagonal (higher weight for more confused classes)
    class_accuracies = np.diag(class_confusion_norm)
    class_weights = 1 - class_accuracies  # Classes with lower accuracy get higher weight
    class_weights = class_weights / class_weights.sum() * 10.0  # Normalize and scale

    print("\n=== Class Weights for Weighted Training ===")
    for i, w in enumerate(class_weights):
        print(f"Class {i}: {w:.4f}")

    # Convert to PyTorch tensor
    class_weights_tensor = torch.FloatTensor(class_weights).to(config.device)

    # Define weighted loss function
    weighted_criterion = nn.CrossEntropyLoss(weight=class_weights_tensor)

    # Train with weighted loss
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    # Training loop
    print("\n=== Training with Class Weighting ===")
    model.train()
    train_losses = []

    for epoch in range(15):
        running_loss = 0.0
        correct = 0
        total = 0

        for inputs, targets in tqdm(train_loader, desc=f"Epoch {epoch+1}/15"):
            inputs, targets = inputs.to(config.device), targets.to(config.device)

            optimizer.zero_grad()
            outputs = model(inputs)
            loss = weighted_criterion(outputs, targets)
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * inputs.size(0)
            _, predicted = torch.max(outputs, 1)
            total += targets.size(0)
            correct += (predicted == targets).sum().item()

        epoch_loss = running_loss / total
        epoch_acc = correct / total
        train_losses.append(epoch_loss)

        print(f"Epoch {epoch+1}/15, Loss: {epoch_loss:.4f}, Acc: {epoch_acc:.4f}")

    # Evaluate on test set
    model.eval()
    test_correct = 0
    test_total = 0
    all_preds = []
    all_targets = []

    with torch.no_grad():
        for inputs, targets in tqdm(test_loader, desc="Testing"):
            inputs, targets = inputs.to(config.device), targets.to(config.device)
            outputs = model(inputs)
            _, predicted = torch.max(outputs, 1)
            test_total += targets.size(0)
            test_correct += (predicted == targets).sum().item()

            all_preds.extend(predicted.cpu().numpy())
            all_targets.extend(targets.cpu().numpy())

    test_acc = test_correct / test_total
    print(f"\nTest Accuracy with Weighted Training: {test_acc:.4f}")

    # Compute confusion matrix
    cm = confusion_matrix(all_targets, all_preds)
    cm_norm = cm / cm.sum(axis=1, keepdims=True)

    # Compare with class weighting
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 7))

    sns.heatmap(class_confusion_norm, annot=True, fmt='.2f', cmap='Blues', ax=ax1)
    ax1.set_title('Confusion Matrix Before Class Weighting')
    ax1.set_xlabel('Predicted Label')
    ax1.set_ylabel('True Label')

    sns.heatmap(cm_norm, annot=True, fmt='.2f', cmap='Blues', ax=ax2)
    ax2.set_title('Confusion Matrix After Class Weighting')
    ax2.set_xlabel('Predicted Label')
    ax2.set_ylabel('True Label')

    plt.tight_layout()
    fig.savefig('weighted_training_comparison.png')

    return {
        'model': model,
        'test_accuracy': test_acc,
        'class_weights': class_weights,
        'confusion_matrix': cm
    }

# Additional refinements: Focal Loss
class FocalLoss(nn.Module):
    def __init__(self, alpha=1, gamma=2, reduction='mean'):
        super(FocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.reduction = reduction
        self.ce_loss = nn.CrossEntropyLoss(reduction='none')

    def forward(self, inputs, targets):
        ce_loss = self.ce_loss(inputs, targets)
        pt = torch.exp(-ce_loss)
        loss = self.alpha * (1 - pt) ** self.gamma * ce_loss

        if self.reduction == 'mean':
            return loss.mean()
        elif self.reduction == 'sum':
            return loss.sum()
        else:
            return loss

def focal_loss_experiment():
    """
    Experiment with Focal Loss to focus more on difficult examples
    """
    config = Config()

    # Load data
    data_loader = MNISTDataLoader(config.batch_size)
    train_loader, val_loader, test_loader = data_loader.load_data()

    # Define model architecture
    input_size = 28 * 28
    hidden_layers = [512, 256, 128, 64]
    layer_sizes = [input_size] + hidden_layers + [10]

    model = MLP(layer_sizes)
    model.to(config.device)

    # Define Focal Loss
    focal_loss = FocalLoss(gamma=2.0)
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    # Training loop
    print("\n=== Training with Focal Loss ===")
    model.train()
    train_losses = []

    for epoch in range(15):
        running_loss = 0.0
        correct = 0
        total = 0

        for inputs, targets in tqdm(train_loader, desc=f"Epoch {epoch+1}/15"):
            inputs, targets = inputs.to(config.device), targets.to(config.device)

            optimizer.zero_grad()
            outputs = model(inputs)
            loss = focal_loss(outputs, targets)
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * inputs.size(0)
            _, predicted = torch.max(outputs, 1)
            total += targets.size(0)
            correct += (predicted == targets).sum().item()

        epoch_loss = running_loss / total
        epoch_acc = correct / total
        train_losses.append(epoch_loss)

        print(f"Epoch {epoch+1}/15, Loss: {epoch_loss:.4f}, Acc: {epoch_acc:.4f}")

    # Evaluate on test set
    model.eval()
    test_correct = 0
    test_total = 0
    all_preds = []
    all_targets = []

    with torch.no_grad():
        for inputs, targets in tqdm(test_loader, desc="Testing"):
            inputs, targets = inputs.to(config.device), targets.to(config.device)
            outputs = model(inputs)
            _, predicted = torch.max(outputs, 1)
            test_total += targets.size(0)
            test_correct += (predicted == targets).sum().item()

            all_preds.extend(predicted.cpu().numpy())
            all_targets.extend(targets.cpu().numpy())

    test_acc = test_correct / test_total
    print(f"\nTest Accuracy with Focal Loss: {test_acc:.4f}")

    # Compute confusion matrix
    cm = confusion_matrix(all_targets, all_preds)

    # Analyze difficult examples
    analyzer = ModelAnalyzer(model, test_loader, config.device)
    misclass_inputs, misclass_targets, misclass_preds = analyzer.get_misclassified_samples(limit=10)

    # Visualize misclassified examples
    fig = Visualizer.plot_misclassified_examples(misclass_inputs, misclass_targets, misclass_preds)
    fig.savefig('focal_loss_misclassified.png')

    return {
        'model': model,
        'test_accuracy': test_acc,
        'confusion_matrix': cm
    }

# Final comprehensive analysis and report
def generate_comprehensive_report():
    """
    Generate a comprehensive report with all experiments and analysis
    """
    # Run all experiments
    config = Config()
    manager = ExperimentManager(config)

    # Standard MLP experiments
    experiments = {
        'small_mlp': {
            'hidden_layers': [128, 64],
            'dropout_rate': 0.2,
            'learning_rate': 0.001,
            'weight_decay': 0.0001,
            'epochs': 15
        },
        'medium_mlp': {
            'hidden_layers': [256, 128, 64],
            'dropout_rate': 0.3,
            'learning_rate': 0.001,
            'weight_decay': 0.0001,
            'epochs': 15
        },
        'large_mlp': {
            'hidden_layers': [512, 256, 128, 64],
            'dropout_rate': 0.4,
            'learning_rate': 0.001,
            'weight_decay': 0.0001,
            'epochs': 15
        }
    }

    # Run standard experiments
    for exp_name, params in experiments.items():
        manager.run_experiment(params, exp_name)

    # Analyze best model
    best_model_analysis = manager.analyze_best_model()

    # Run weighted training experiment
    weighted_results = weighted_training_experiment()

    # Run focal loss experiment
    focal_results = focal_loss_experiment()

    # Advanced analysis
    advanced_analysis_results = advanced_analysis()

    # Generate report
    print("\n=== COMPREHENSIVE REPORT ===")
    print("\n1. Standard MLP Experiments")
    for exp_name, results in manager.results.items():
        print(f"\n{exp_name}:")
        print(f"  Architecture: {results['model_params']['hidden_layers']}")
        print(f"  Dropout: {results['model_params']['dropout_rate']}")
        print(f"  Test Accuracy: {results['test_metrics']['accuracy']:.4f}")
        print(f"  Test F1 Score: {results['test_metrics']['f1']:.4f}")

    print("\n2. Best Model Analysis")
    print(f"  Best Model: {best_model_analysis['best_exp_name']}")
    print(f"  Test Accuracy: {best_model_analysis['test_metrics']['accuracy']:.4f}")

    print("\n3. Weighted Training Results")
    print(f"  Test Accuracy: {weighted_results['test_accuracy']:.4f}")
    print("  Class Weights:")
    for i, w in enumerate(weighted_results['class_weights']):
        print(f"    Class {i}: {w:.4f}")

    print("\n4. Focal Loss Results")
    print(f"  Test Accuracy: {focal_results['test_accuracy']:.4f}")

    print("\n5. Advanced Analysis")
    print("  Most Common Confusion Pairs:")
    for (true, pred), count in advanced_analysis_results['top_confusions']:
        print(f"    True: {true}, Predicted: {pred} - Count: {count}")

    print("\n6. Key Interpretations and Findings:")
    print("  - The model struggles most with distinguishing visually similar digits")
    print("  - Certain writing styles cause consistent misclassifications")
    print("  - Weighted training helped improve performance on difficult classes")
    print("  - Focal loss provided better focus on hard examples during training")
    print("  - The model's decision boundaries are clearest between visually distinct digits")

    print("\n7. Suggested Improvements:")
    print("  - Use convolutional layers to better capture spatial features")
    print("  - Implement data augmentation to improve generalization")
    print("  - Try ensemble methods to combine multiple models")
    print("  - Apply more advanced regularization techniques")
    print("  - Explore transfer learning from pre-trained models")

    return {
        'standard_results': manager.results,
        'best_model': best_model_analysis,
        'weighted_results': weighted_results,
        'focal_results': focal_results,
        'advanced_analysis': advanced_analysis_results
    }

# Run the main function if this script is executed
if __name__ == "__main__":
    main()
```
